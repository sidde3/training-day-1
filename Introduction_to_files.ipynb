## Validating file existance using Pathlib

```python
from pathlib import Path

file = Path("info.txt")

if file.exists():
    print("File exists")
    if file.is_file():
        print("Valid file exists")
else:
    print("File does not exist")
```

## Reading the fist line of a file

```python
with open("info.txt", "r") as f:
    first_line = f.readline()
    print(first_line)
```

## Read a specific line

```python
with open("info.txt", "r") as f:
    lines = f.readlines()
    print(lines[1])
```

# Example of downloading files from S3 bucket

```python
import os
import boto3
from botocore.client import Config
from botocore.exceptions import ClientError

# Initialize S3 client
s3_client = boto3.client(
    's3',
    endpoint_url=os.getenv('AWS_S3_ENDPOINT'),
    aws_access_key_id=os.getenv('AWS_ACCESS_KEY_ID'),
    aws_secret_access_key=os.getenv('AWS_SECRET_ACCESS_KEY', 'ChangeMe-MinIO-2026'),
    config=Config(signature_version='s3v4')
)
```

## List all buckets


```python
try:
    response = s3_client.list_buckets()
    print("Available buckets:")
    for bucket in response['Buckets']:
        print(f" - {bucket['Name']}")
except ClientError as e:
    print("Error listing buckets:", e)
    exit()
```

## Choose a bucket

```python
bucket_name = 'training'
print(f"\nListing files in bucket: {bucket_name}")
```

## List and download files


```python
paginator = s3_client.get_paginator('list_objects_v2')

try:
    for page in paginator.paginate(Bucket=bucket_name):
        contents = page.get('Contents', [])

        if not contents:
            print("No files found in bucket.")
            continue

        for obj in contents:
            key = obj['Key']
            print(f"Found file: {key}")

except ClientError as e:
    print("Error listing objects:", e)
```

```python
key = 'jwks.json'
try:
    print(f"Downloading {key}...")
    s3_client.download_file(bucket_name, key, key)
    print(f"Downloaded: {key}\n")
except ClientError as e:
    print(f"Failed to download {key}: {e}")
```
