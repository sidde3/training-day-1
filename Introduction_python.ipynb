## Wrapper function

```python
def say_hello():
    print("Hello!")

def wrapper(func):
    def inner():
        print("Before function call")
        func()
        print("After function call")
    return inner

wrapped = wrapper(say_hello)
wrapped()
```

## Decorator function

```python
def wrapper(func):
    def inner():
        print("Before function call")
        func()
        print("After function call")
    return inner

@wrapper
def say_hello():
    print("Hello!")

say_hello()
```

## Common Problem

```python
def wrapper(func):
    def inner(*args, **kwargs):
        print("Before")
        result = func(*args, **kwargs)
        print("After")
        return result
    return inner

@wrapper
def add(a, b):
    return a + b

print(add(2, 3))
```

```python
import time

def timer(func):
    def inner(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        print("Time:", time.time() - start)
        return result
    return inner


@timer
def add(a, b):
    time.sleep(1)
    return a + b

print(add(10, 20))
```

## Manual Wrapping

```python
def add(a, b):
    time.sleep(1)
    return a + b

wrapped_add = timer(add)

print(wrapped_add(10, 20))
```

```python
def print_details(**kwargs):
    print(kwargs)

print_details(name="John", age=25)
```

## 3D plotting (mpl_toolkits)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  

# Sample 3D data
x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])
z = np.array([2, 4, 6, 8])

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot a 3D line
ax.plot(x, y, z, label='3D line', marker='o')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Simple 3D Line Plot')
ax.legend()

plt.show()
```

## Check the available memory in GPU

## CPU VS GPU

```python
import torch
import time

# Size of the tensor
N = 5000

# --- CPU computation ---
x_cpu = torch.randn(N, N)
y_cpu = torch.randn(N, N)

start = time.time()
z_cpu = torch.matmul(x_cpu, y_cpu)
end = time.time()

print(f"CPU time: {end - start:.4f} seconds")

# --- GPU computation ---
x_gpu = x_cpu.to('cuda')
y_gpu = y_cpu.to('cuda')

torch.cuda.synchronize()  # Wait for GPU to finish previous ops
start = time.time()
z_gpu = torch.matmul(x_gpu, y_gpu)
torch.cuda.synchronize()  # Wait for GPU to finish
end = time.time()

print(f"GPU time: {end - start:.4f} seconds")
```
